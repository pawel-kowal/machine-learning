###Przygotwanie danych

```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
np.set_printoptions(precision = 6, suppress=True, edgeitems=30, linewidth=120,
                    formatter=dict(float=lambda x: f'{x:.2f}'))
sns.set(font_scale=1.3)
sklearn.__version__
```

```python
from sklearn.datasets import make_regression

data, target = make_regression(n_samples=100, n_features=1, n_targets=1,
                               noise=30.0, random_state=42)

print(f'data shape: {data.shape}')
print(f'target shape: {target.shape}')
```

```python
data[:5]
```

```python
target[:5]
```

```python
plt.figure(figsize=(8,6))
plt.title('Regresja liniowa')
plt.xlabel('cecha x')
plt.ylabel('zmienna docelowa')
plt.scatter(data, target, label='cecha x')
plt.legend()
plt.plot()
```

###Regresja liniowa przy użyciu scikit-learn

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(data, target)
```

```python
#metoda score() dokonuje oceny modelu
regressor.score(data, target)
```

```python
#metoda predict() dokonuje predykcji na podstawie modelu
y_pred = regressor.predict(data)
y_pred
```

###Wizualizacja graficzna modelu

```python
plt.figure(figsize=(8,6))
plt.title('Regresja liniowa')
plt.xlabel('cecha x')
plt.ylabel('zmienna docelowa')
plt.scatter(data, target, label='cecha x')
plt.plot(data, y_pred, color='red', label='model')
plt.legend()
plt.show()
```

###Współczynnik R2
R2 = 1-(u/v), gdzie u to:

```python
((y_true-y_pred)**2).sum()
```

oraz v:

```python
((y_true-y_true.mean())**2).sum()
```

czyli:

```python
1-(((y_true-y_pred)**2).sum()/((y_true-y_true.mean())**2).sum())
```

Najlepiej, gdy R2 jest bliskie jeden. R2 może być ujemne.

###Regresja z podziałem na zbiót testowy i treningowy

```python
data, target = make_regression(n_samples=1000, n_features=1, n_targets=1, noise=15.0, random_state=42)

print(f'data shape: {data.shape}')
print(f'target shape: {target.shape}')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.25)

print(f'X_train shape {X_train.shape}')
print(f'X_test shape {X_test.shape}')
print(f'y_train shape {y_train.shape}')
print(f'y_test shape {y_test.shape}')
```

```python
plt.figure(figsize=(8,6))
plt.title('Regresja liniowa train vs test')
plt.xlabel('cecha x')
plt.ylabel('zmienna docelowa')
plt.scatter(X_train, y_train, label='zbiór treningowy', color='gray', alpha=0.5)
plt.scatter(X_test, y_test, label='zbiór trestowy', color='gold', alpha=0.5)
plt.legend()
plt.plot()
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
regressor.score(X_train, y_train)
```

```python
regressor.score(X_test, y_test)
```

###Regresja liniowa - zbiór treningowy - wizualizacja

```python
plt.figure(figsize=(8,6))
plt.title(f'Regresja liniowa: zbiór treningowy\nY={round(regressor.intercept_,2)}+{round(regressor.coef_[0],2)}X1')
plt.xlabel('cecha x')
plt.ylabel('zmienna docelowa')
plt.scatter(X_train, y_train, label='zbiór treningowy', color='gray', alpha=0.5)
plt.plot(X_train, regressor.intercept_+regressor.coef_[0]*X_train, color='red')
plt.legend()
plt.show()
```

###Regresja liniowa - zbiór testowy - wizualizacja

```python
plt.figure(figsize=(8,6))
plt.title(f'Regresja liniowa: zbiór treningowy\nY={round(regressor.intercept_,2)}+{round(regressor.coef_[0],2)}X1')
plt.xlabel('cecha x')
plt.ylabel('zmienna docelowa')
plt.scatter(X_test, y_test, label='zbiór treningowy', color='gold', alpha=0.5)
plt.plot(X_test, regressor.intercept_+regressor.coef_[0]*X_test, color='red')
plt.legend()
plt.show()
```

###Predykcja na podstawie modelu

```python
y_pred = regressor.predict(X_test)

predictions = pd.DataFrame(data={'y_true':y_test, 'y_pred':y_pred})
predictions.head()
```

```python
predictions['error'] = predictions['y_true'] - predictions['y_pred']
predictions.head()
```
